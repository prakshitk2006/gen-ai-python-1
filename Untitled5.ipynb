```python
#default argument--these argument value given during defined a function
def area_perimeter(width = 2, height = 4):
  area = width*height
  perimeter = 2*(width+height)
  return area, perimeter
```

```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
  return sum
```

```python
#keyword variable length argument
#keyword veriable length argument behave a dictionary
def test(**kwargs):
    print(kwargs)
    print(len(kwargs))
    print(type(kwargs))
```

```python
def pass_students(**kwargs):
  pass_student_list = []
  for name, marks in kwargs.items():
    if marks>33:
      pass_student_list.append(name)
  return pass_student_list
```

```python
pass_students(A=40, B=30, C=60)
```
